# Passo a passo do problema

- Passo 1: buscas de dados das ações automaticamente
- Passo 2: realizar as analises solicitadas pelo gestor
- Passo 3: enviar automaticamente o email com o resultado da análise

```python
import yfinance as yf

ticker = input("Digite o codigo da ação: ")
dados = yf.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()

maxima = round(fechamento.max(), 2)
minima = round(fechamento.min(), 2)
atual = round(fechamento[-1], 2)

print(maxima)
print(minima)
print(atual)

import pyautogui
import pyperclip

#pausa de 2 segundos
pyautogui.PAUSE = 4

# abrir uma nova aba(ctrl + t)
pyautogui.hotkey("ctrl", "t")

#digitar www.gmail.com e dar enter
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# clicar no botão Escrever
pyautogui.click(x=67, y=168)

#digitar o destinatario(tab)
pyperclip.copy("eduardoflorentinoramos@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#digitar o assunto(tab)
pyperclip.copy("Analises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

#preencher o corpo do email

mensagem = f"""
Prezado gestor,

Seguem conforme solicitado, as analises do ultimos 6 meses da ação {ticker}:

Cotação máxima: R$ {maxima}
Cotação minima: R$ {minima}
Cotação atual: R$ {atual}

Qualquer dúvida, fico a disposição!

Att.
Eduardo
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

pyautogui.click(x=840, y=635)
print("E-mail enviado com Sucesso!")

"""import time

time.sleep(7)
pyautogui.position()"""
```
